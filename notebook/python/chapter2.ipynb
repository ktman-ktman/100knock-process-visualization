```python
def merge(idx_t: tuple) -> str:
    col = ""
    for idx_i in idx_t:
        if "Unnamed" in idx_i:
            continue
        elif idx_i.startswith("〔"):
            col += f"{idx_i[1:-1]}_"
        else:
            col += f"{idx_i}_"

    if col[-1] == "_":
        col = col[:-1]
    return col
```

```python
with pd.ExcelFile("../../data/chapter-2/data/1-2-2020.xlsx") as xlsx:
    sheet_l = xlsx.sheet_names
    sheet_l.sort()
    result_l = list()
    for sheet_i in sheet_l:
        # read file
        tmp_df = pd.read_excel("../../data/chapter-2/data/1-2-2020.xlsx", skiprows=1, header=[0,1,2], skipfooter=4, sheet_name=sheet_i)
        tmp_df.columns = [merge(x) for x in df.columns]
        tmp_df.loc[:, "年月"] = sheet_i
        result_l.append(tmp_df)

    df = pd.concat(result_l, axis="index", ignore_index=True)
```
